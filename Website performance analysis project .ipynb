```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data-export.csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df.head()
```

```python
df.columns = df.iloc[0]
df= df.drop(index = 0).reset_index(drop = True)
df.columns = ["channel group", "DateHour", "Users", "Sessions", "Engaged sessions", "Average engagement time per session", "Engaged sessions per user", "Events per session", "Engagement rate	", "Event count"]
```

```python
df.info()
```

```python
df.describe()
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors="coerce")
df.head()
```

```python
numeric_cols = df.columns.drop(["channel group", "DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df["hour"] = df["DateHour"].dt.hour
```

## sessions and user over time

```python
sns.set(style = 'whitegrid')
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions", "Users"]].sum().plot(ax=plt.gca())
plt.xlabel("DateHour")
plt.ylabel("count")
plt.show()
```

## Total Users by Channel

```python
plt.figure(figsize=(8,5))
sns.barplot(data =df, x="channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title("Total user by channel")
plt.xticks(rotation=45)
plt.show()
```

## average engagement Time

```python
plt.figure(figsize=(8,5))
sns.barplot(data =df, x="channel group", y="Average engagement time per session", estimator=np.mean, palette="magma")
plt.title("Avg engagemnt time by channel")
plt.xticks(rotation=45)
plt.show()
```

## Engagement rate of distribution by channel

```python
plt.figure(figsize=(8,5))
sns.boxplot(data =df, x="channel group", y="Engagement rate\t", palette="coolwarm")
plt.title("Avg engagemnt rate distribution by channel")
plt.xticks(rotation=45)
plt.show()
```

## Engagae vs Non Engage Sessions

```python
df.head()
```

```python
# Clean column names first
df.columns = df.columns.str.replace("\t", "", regex=False).str.strip()

# Grouped session dataframe
session_df = df.groupby("channel group")[["Sessions","Engaged sessions"]].sum().reset_index()

# Fix column names again just to be safe
session_df.columns = session_df.columns.str.replace("\t", "", regex=False).str.strip()

# Create non-engaged column
session_df["non-engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]

# Melt the dataframe correctly
session_df_melted = session_df.melt(
    id_vars="channel group",
    value_vars=["Engaged sessions", "non-engaged"],
    var_name="variable",
    value_name="value"
)

# Plot using melted dataframe
plt.figure(figsize=(8,5))
sns.boxplot(data=session_df_melted, x="channel group", y="value", hue="variable")
plt.title("Engaged Vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

## Traffic by hour and Channel

```python
heatmap_data = df.groupby(["hour", "channel group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu",linewidth=.5, annot=True, fmt ='.0f')
plt.title("Traffic by hour and channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of day")
plt.show()
```

## Engagement rate Vs sessions over time

```python
df_plot = df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.legend()
plt.title("Engagement rate vs sessions over time")
plt.xlabel("DateHour")
plt.grid(True)
plt.show()
```
